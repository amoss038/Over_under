# Setting up the Chromedriver and setting the NBA.COM URL for webscraping.

```python
#url for current season
url_current = 'https://www.nba.com/stats/teams/boxscores/'

#url for last season
url_2019_2020 = 'https://www.nba.com/stats/teams/boxscores/?Season=2019-20&SeasonType=Regular%20Season'

#url for 2 seasons ago
url_ago_2018_2019 = 'https://www.nba.com/stats/teams/boxscores/?Season=2018-19&SeasonType=Regular%20Season'

#url for 3 seasons ago
url_ago_2017_2018 = 'https://www.nba.com/stats/teams/boxscores/?Season=2017-18&SeasonType=Regular%20Season'
```

# function for webscraping the NBA website game log stats

```python
def web_scraper(website):
    #setting the url and accessing the url with the webdriver for season
    driver = webdriver.Chrome()
    driver.get(website)
    
    #selecting the xpath of the selector element on nba website
    select = Select(driver.find_element_by_xpath(r'/html/body/main/div/div/div[2]/div/div/nba-stat-table/div[1]/div/div/select'))
    
    #selecting by index 0 because ALL is at the first index
    select.select_by_index(0)
    
    #parsing out the table and accessing the class and table -->'nba-stat-table__overflow'
    src =driver.page_source
    parser = BeautifulSoup(src, "lxml")
    table = parser.find('div', attrs = {'class': 'nba-stat-table__overflow'})

    # creating a header list of table headers
    headers = table.findAll('th')
    headerlist = [h.text.strip() for h in headers]
    
    #creating the rows object for the eventual DF
    rows = table.findAll('tr')
    
    #Extracting the text from the HTML object 'tr' and creating a list called game stats that 
    #includes the data for each column
    game_stats = [[td.getText().strip() for td in rows[i].findAll('td')] for i in range(len(rows))]
    
    #converting gamestate and headerlist into a DF
    games_df = pd.DataFrame(game_stats, columns = headerlist)
    
    #Dropping the season column because it was a hidden column that did not include any useful information
    games_df.drop('Season', axis =1, inplace =True)
    
    #dropping the first row of the DF because it only contained N/A values.
    games_df_final = games_df.iloc[1:]
    
    return games_df_final
```

```python
game_logs_2017_2018 = web_scraper(url_ago_2017_2018)
```
